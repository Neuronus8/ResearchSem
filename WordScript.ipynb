```python
import collections
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
file = open('Friends_script.csv', encoding="utf8")
a= file.read()
replwords = ['you','i','the','to','a','just','my','this','me','of', ".", ",", ":", "\"", "!", "*", "â€œ", "â€˜"]
wordcount = {}
```

```python
Totalwords = a.lower().split()
for word in Totalwords:
    if word not in replwords:
        if word not in wordcount:
            wordcount[word] = 1
        else:
            wordcount[word] += 1
    else:
        if word not in wordcount:
            wordcount[word] = 0
        else:
            wordcount[word] += 0
```

```python
#n_print = int(input("How many most common words to print: "))
n_print = 10
print("\nOK. The {} most common words are as follows\n".format(n_print))
word_counter = collections.Counter(wordcount)
```

```python
for word, count in word_counter.most_common(n_print):
    print(word, ": ", count)# Close the file
```

```python
file.close()# Create a data frame of the most common words 
# Draw a bar chart
lst = word_counter.most_common(n_print)
df = pd.DataFrame(lst, columns = ['Word', 'Count'])
df.plot.bar(x='Word',y='Count')
```
